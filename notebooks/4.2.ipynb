```python
# 4.2

from scipy.stats import beta, gamma
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Log-Odds Example
a = 1
b = 1

bdist_prior = beta(a=a, b=b)

theta_prior_mc = bdist_prior.rvs(10000)
gamma_prior_mc = np.log(theta_prior_mc / (1 - theta_prior_mc))
```

```python
n0 = 860 - 441
n1 = 441

bdist_post = beta(a=a+n1, b=b+n0)

theta_post_mc = bdist_post.rvs(10000)
gamma_post_mc = np.log(theta_post_mc / (1 - theta_post_mc))
```

```python
sns.distplot(gamma_prior_mc, hist=False, kde=True, 
             color = 'darkblue', 
             hist_kws={'edgecolor':'darkblue'},
             kde_kws={'linewidth': 2});
plt.xlim(-4.5, 4.5);
plt.xlabel('\u03B3');
plt.ylabel('p(\u03B3)');
```

```python
sns.distplot(gamma_prior_mc, hist=False, kde=True, 
             color = 'darkblue', 
             hist_kws={'edgecolor':'darkblue'},
             kde_kws={'linewidth': 2});
sns.distplot(gamma_post_mc, hist=False, kde=True, 
             color = 'green', 
             hist_kws={'edgecolor':'green'},
             kde_kws={'linewidth': 2});
plt.xlim(-4.5, 4.5);
plt.xlabel('\u03B3');
plt.ylabel('p(\u03B3|y1, y2,...yn)');
```

```python
# Functions of Two Parameters

a = 2
b = 1

sy1 = 217
n1 = 111
sy2 = 66
n2 = 44

shape1 = a + sy1
rate1 = b + n1
scale1 = 1 / rate1

shape2 = a + sy2
rate2 = b + n2
scale2 = 1 / rate2

theta1_mc = gamma(a=shape1, scale=scale1).rvs(10000)
theta2_mc = gamma(a=shape2, scale=scale2).rvs(10000)
```

```python
np.mean(theta1_mc > theta2_mc)
```

```python
sns.distplot(theta1_mc/theta2_mc, hist=False, kde=True, 
             color = 'darkblue', 
             hist_kws={'edgecolor':'darkblue'},
             kde_kws={'linewidth': 2});
plt.xlim(0.5, 2.5);
plt.xlabel('\u03B3=\u03B81/\u03B82');
plt.ylabel('p(\u03B3|y1, y2,...yn)');
```
